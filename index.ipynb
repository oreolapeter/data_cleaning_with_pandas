```python
# creates a function that extract the state and city from the destination column

def extract_strings(text):
    result = text.split(',')[1]
    return result

def extract_strings_2(text_2):
    result_2 = text_2.split(',')[0]
    return result_2
```

```python
travel_dataset['State'] = travel_dataset['Destination'].apply(extract_strings) #extracts the state
travel_dataset['City'] = travel_dataset['Destination'].apply(extract_strings_2) #extract the city
travel_dataset.head(3)
```

```python
# function allows us to extract string separated by space

def extract_strings_space(text):
    result_space = text.split(' ')[0]
    return result_space

def extract_strings_space1(text):
    result_space2 = text.split(' ')[-1]
    return result_space2
```

```python
travel_dataset['First Name'] = travel_dataset['Employee'].apply(extract_strings_space) #extract the first name

travel_dataset['Last Name'] = travel_dataset['Employee'].apply(extract_strings_space1) #extract the last name

travel_dataset #view the dataset to see the state after cleaning
```
